```python
import sqlite3
database = 'database.sqlite'
conn = sqlite3.connect(database)
print("done")
```

```python
conn.execute("""CREATE TABLE CLASS_10
                (SNO INT PRIMARY KEY NOT NULL,
                Roll_No INT NOT NULL,
                NAME TEXT NOT NULL,
                AGE INT DEFAULT (15),
                Gender TEXT NOT NULL,
                Email_ID TEXT NOT NULL,
                Contact_No REAL NOT NULL);""")
print("TABLE CREATED SUCCESFULLY")
```

```python
conn.execute("INSERT INTO CLASS_10 (SNO,Roll_No,NAME,AGE,Gender,Email_ID,Contact_No) \
  VALUES (1, 1, 'allen', 14, 'male', 'allen@gmail.co.uk', 8080900)");

conn.execute("INSERT INTO CLASS_10 (SNO,Roll_No,NAME,AGE,Gender,Email_ID,Contact_No) \
  VALUES (2, 2, 'aisha', 14, 'female', 'aish@gmail.co.uk', 9080900)");

conn.commit()
print("records created succesfully")
```
